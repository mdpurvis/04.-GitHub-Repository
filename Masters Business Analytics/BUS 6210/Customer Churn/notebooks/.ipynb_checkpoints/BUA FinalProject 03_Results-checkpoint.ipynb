```python
# EDA - Data Review
import pandas as pd
import numpy as np
import missingno
from collections import Counter
import math

# EDA - Data Visualizations
import matplotlib.pyplot as plt
import seaborn as sns
from pylab import rcParams
%matplotlib inline

# Remove warnings
import warnings
warnings.filterwarnings('ignore')


pd.set_option('display.max_rows', 90)
pd.set_option('display.max_columns', 200)
pd.set_option('display.width', 1000)
```

```python
results = pd.read_csv('../data/final/cell2cell_results.csv', index_col='CustomerID')
test = pd.read_csv('../data/raw/cell2cellholdout.csv', index_col='CustomerID').drop(columns='Churn')
print('Shape of Prediction Dataset: ', results.shape)
print('Shape of Prediction Dataset: ', test.shape)
```

```python
df = test.join(results).dropna()
df.Churn.replace(to_replace=1, value='Yes', inplace=True)
df.Churn.replace(to_replace=0, value='No', inplace=True)
df.shape
```

```python
df
```

```python
# Data to plot for the % of target variable
sns.set(font_scale = 1.7)
labels =df['Churn'].value_counts(sort = True).index
sizes = df['Churn'].value_counts(sort = True)
colors = ["lightsteelblue","salmon"]
explode = (0.1,0)  # explode 1st slice
 
rcParams['figure.figsize'] = 8,8
# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=270,)
plt.title('Percent of Customers Who Are Predicted to Churn')
plt.show()
```
