```python
try:
    import pdfplumber
    import re
    import pandas as pd
except ImportError:
    input('Package failed to import. Click enter to terminate: ')  
```

```python
pdf_file = ['../data/raw/2020-03-31 Allstate Transcript.pdf', '../data/raw/2019-12-31 Allstate Transcript.pdf']

all_text = ['df', 'df2']

for ea in range(len(pdf_file)):
    with pdfplumber.open(pdf_file[ea]) as pdf:
        print(pdf_file[ea])
        all_text[ea] = ''
        # page = pdf.pages[0] - comment out or remove line
        # text = page.extract_text() - comment out or remove line
        for pdf_page in pdf.pages:
           single_page_text = pdf_page.extract_text()
    #        print( single_page_text )
           # separate each page's text with newline
           all_text[ea] = all_text[ea] + '\n' + single_page_text 
           all_text[ea] = all_text[ea].replace(',','') 
        
        
all_text
```
